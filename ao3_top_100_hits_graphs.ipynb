```python
# Import necessary modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from turtle import color
import datetime
from datetime import datetime
```

```python
# Import the csv file scraped from the website
ao3_top_100_hits = pd.read_csv('ao3_top_100_hits.csv')
```

```python
# Cleaning up the freeform tag data
freeform_tags = []
for i in range(100):
    temp_tags = ao3_top_100_hits.freeforms[i].lstrip("[")
    temp_tags = temp_tags.rstrip("]")
    temp_tags = temp_tags.split(",")
    for elm in temp_tags:
        elm2 = elm.strip()
        elm3 = elm2.strip('''"''')
        elm4 = elm3.strip("'")
        freeform_tags.append(elm4)

count_of_freeform_tags = {}

for i in freeform_tags:
    count_of_freeform_tags [i] = freeform_tags.count(i)
    
# print(count_of_freeform_tags)
```

```python
# Getting the most popular freeform tags
top_freeform_tags = {}

top_freeforms_ordered = {}

top_keys = [key for key, value in count_of_freeform_tags.items() if value >= 6]

for i in top_keys:
    top_freeform_tags[i] = count_of_freeform_tags[i]

in_order_freeform = sorted(top_freeform_tags, key=top_freeform_tags.get, reverse=True)

for i in in_order_freeform:
    top_freeforms_ordered[i] = count_of_freeform_tags[i]
    
valdel = top_freeforms_ordered.pop('No Freeform Tags')
print(top_freeforms_ordered)
```

```python
# Graphing the most popular tags
freeform_names = list(top_freeforms_ordered.keys())
freeform_values = list(top_freeforms_ordered.values())

plt.barh(range(len(top_freeforms_ordered)), freeform_values, tick_label=freeform_names, color = 'green')
plt.title('Occurences of Most Popular Freeform Tags Used in Top 100 Visited Fanfictions on AO3')
plt.xlabel('Number of Occurences')
# plt.ylabel('Freeform Tags')
plt.show()
```

```python
#Set up the variables
x_values = list(range(len(top_freeforms_ordered)))
y_values = freeform_values
x_labels = freeform_names

# print(x_values)
# print(y_values)
# print(x_labels)

# plt.figure(figsize=(9,5))
fig, ax = plt.subplots( nrows=1, ncols=1, figsize=(12,5) )
sns.barplot(y=x_values, x=y_values, palette='rocket', orient='h')
ax.set_yticks(x_values)
ax.set_yticklabels(x_labels)
ax.tick_params(labelsize=10.5)
plt.title('Occurences of Most Popular Freeform Tags Used in Top 100 Visited Fanfictions on AO3', fontsize=14)
plt.xlabel('Number of Occurences', fontsize=12)
plt.show()
# fig.savefig('freeform_tags_barplot.png')
plt.clf()
```

```python
# Investigating the prevalence of slash_categories
# print(ao3_top_100_hits.slash_categories)

# Cleaning up the data:
slash_cats = []
for i in range(100):
    temp_cats = ao3_top_100_hits.slash_categories[i].lstrip("[")
    temp_cats = temp_cats.rstrip("]")
    temp_cats = temp_cats.split(",")
    for elm in temp_cats:
        elm2 = elm.strip()
        elm3 = elm2.strip("'")
        slash_cats.append(elm3)

print(slash_cats)
count_of_slash_cats_unordered = {}

for i in slash_cats:
    count_of_slash_cats_unordered [i] = slash_cats.count(i)

sorted_slash_cats = sorted(count_of_slash_cats_unordered, key=count_of_slash_cats_unordered.get, reverse=True)

count_of_slash_cats = {}
for elm in sorted_slash_cats:
    count_of_slash_cats[elm] = count_of_slash_cats_unordered[elm]

print(count_of_slash_cats)
```

```python
#Strip fandom list down to base components
clean_fandoms_list = []

for row in ao3_top_100_hits.fandoms:
    row_list = []
    temp1 = row.lstrip("[")
    temp2 = temp1.rstrip("]")
    temp3 = temp2.split(",")
    # print("Temp3:", temp3)
    for elm in temp3:
        elm2 = elm.strip()
        elm3 = elm2.strip('''"''')
        elm4 = elm3.strip("'")
        row_list.append(elm4)
        # print("elm4:", elm4)
    clean_fandoms_list.append(row_list)
    # print(row_list)

ao3_top_100_hits["fandoms_clean"] = clean_fandoms_list

# print(ao3_top_100_hits.fandoms_clean[13][1])
#SUCCESS!
```

```python
# Create a list of all the unique fandoms

fandoms_unique = []

for row in ao3_top_100_hits.fandoms_clean:
    # print("row:", row)
    for i in row:
        # print(i in fandoms_unique)
        if (i in fandoms_unique) == True:
            # print(i, "is already there")
            None
        else:
            fandoms_unique.append(i)
            # print(i, 'appended')

# print(fandoms_unique)
```

```python
# Calculating the occurences of each fandom

fandoms_count = {}
for fandom in fandoms_unique:
    fandoms_count[fandom] = 0

for row in ao3_top_100_hits.fandoms_clean:
    for fandom in row:
        fandoms_count[fandom] = (fandoms_count[fandom] + 1)

print((fandoms_count))
```

```python
# Calculating the average kudos of each fandom

fandoms_kudos = {}
for fandom in fandoms_unique:
    fandoms_kudos[fandom] = 0

for fandom in fandoms_kudos:
    for i in range(100):
        if (fandom in ao3_top_100_hits.fandoms_clean[i]) == True:
            fandoms_kudos[fandom] = fandoms_kudos[fandom] + ao3_top_100_hits.kudos[i]

fandoms_avg_kudos = {}
for fandom in fandoms_unique:
    fandoms_avg_kudos[fandom] = ((fandoms_kudos[fandom] / fandoms_count[fandom]) / 1000)

# print(fandoms_avg_kudos)
```

```python
# Calculating the average hits for each fandom

fandoms_hits = {}
for fandom in fandoms_unique:
    fandoms_hits[fandom] = 0

for fandom in fandoms_hits:
    for i in range(100):
        if (fandom in ao3_top_100_hits.fandoms_clean[i]) == True:
            fandoms_hits[fandom] = fandoms_hits[fandom] + ao3_top_100_hits.hits[i]

fandoms_avg_hits = {}
for fandom in fandoms_unique:
    fandoms_avg_hits[fandom] = ((fandoms_hits[fandom] / fandoms_count[fandom]) / 1000000)

# print(fandoms_avg_hits)
```

```python
# creating a dataframe to base the scatterplot off
df_fandoms = pd.DataFrame.from_dict(fandoms_unique)
df_fandoms.rename(columns = {df_fandoms.columns[0]:'Fandom'}, inplace = True)
df_fandoms['Kudos'] = fandoms_avg_kudos.values()
df_fandoms['Hits'] = fandoms_avg_hits.values()
df_fandoms['Count'] = fandoms_count.values()
# print(df_fandoms)
```

```python
#Creating the scatterplot (finally!)

sns.scatterplot(data=df_fandoms, x='Hits', y='Kudos', size='Count', hue='Count')
plt.xlabel('Average Hits (Millions)')
plt.ylabel('Average Kudos (Thousands)')
plt.title('Popularity of Fandoms within the Top 100 Visited Fanfictions on AO3')
plt.annotate('Harry Potter', xy=(1.5, 16))
plt.annotate('Guardians of the Galaxy', xy=(1.0, 120))
plt.annotate('Original - Fandom', xy=(1.87, 4))
plt.show()
plt.clf()
```
